```python
class Animal:
  def __init__(self, name):
    self.name = name


class Dog(Animal):
  def __init__(self, name, breed):
    super().__init__(name)
    self.breed = breed

my_dog = Dog("Duffy", "Goldendoodle")
print(my_dog.name)
```

```python
class Student:
  school_name = "Touro"

  def __init__(self, name):
    self.name = name

  def get_name(self):
    return self.name

  @classmethod
  def get_school(cls):
    return cls.school_name

  @staticmethod
  def is_valid_gpa(gpa : float):
    return 0.0 < gpa < 4.3

student = Student("Name")
print(student.is_valid_gpa(4.0))
print(student.get_school())
```

```python
class Math:
  @staticmethod
  def add(x:int, y:int):
    return x + y

  @staticmethod
  def multiplication(x : int, y : int):
    return x * y

print(Math.add(1, 1))
print(Math.multiplication(1, 1))
```

```python
class GuessValidator:
  @staticmethod
  def is_single_letter(s : str):
    return len(s) == 1

  @staticmethod
  def normalise_guess(s : str):
    return s.lower().strip()

print(GuessValidator.is_single_letter("a"))
print(GuessValidator.is_single_letter("aaa"))

print(GuessValidator.normalise_guess("    A    "))
```

```python
from abc import ABC, abstractmethod
class Animal(ABC):
  def __init__(self, name : str):
    self.name = name

  def get_name(self):
    return self.name

  @abstractmethod
  def speak(self):
    pass

class Cat(Animal):
  def __init__(self, name):
    super().__init__(name)

  def speak(self):
    return "nya"

class Dog(Animal):
  def __init__(self, name):
    super().__init__(name)

  def speak(self):
    return "wan"

the_cat = Cat("nyanyadesu")
print(the_cat.speak())

the_dog = Dog("wanwandesu")
print(the_dog.speak())
```

```python
# Word game abstract class
from abc import ABC, abstractmethod
class WordGameController(ABC):
  @abstractmethod
  def play(self):
    pass

  @abstractmethod
  def make_guess(self, letter : str):
    pass

  @abstractmethod
  def check_game_over(self) -> bool:
    pass

class TinyWordGame(WordGameController):
  def __init__(self, secret_word : str):
    self.secret_word = secret_word
    self.guesses = set()

  def play(self):
    print("Welcome to the tiny word game, guess the secret letter")

  def make_guess(self, letter : str):
    self.guesses.add(letter)
    return letter in self.secret_word

  def check_game_over(self):
    return all(letter in self.guesses for letter in self.secret_word)

game = TinyWordGame("word")
game.play()
game.make_guess("c")
print(game.check_game_over())
```

```python
class HintWordGame(WordGameController):
  def __init__(self, secret_word : str, hint : str):
    self.secret_word = secret_word
    self.hint = hint
    self.guesses = set()

  def play(self):
    print(f"Welcome to the word game. Here is your hint: {self.hint}")

  def make_guess(self, letter : str):
    self.guesses.add(letter)
    return letter in self.secret_word

  def check_game_over(self):
    return all(letter in self.guesses for letter in self.secret_word)

hintgame = HintWordGame("master sword", "The Sword that seals the Darkness")
hintgame.play()
hintgame.make_guess("a")
print(hintgame.check_game_over())

```

```python
class LimitedGuessWordGame(WordGameController):
  def __init__(self, secret_word : str,max_guesses : int):
    self.secret_word = secret_word
    self.max_guesses = max_guesses
    self.num_guesses = 0
    self.guesses = set()

  def play(self):
    print(f"Welcome to limited guess word game. You have {self.max_guesses} maximum guesses.")

  def make_guess(self, letter : str):
    self.guesses.add(letter)
    self.num_guesses += 1
    self.remaining_guesses()
    return letter in self.secret_word

  def remaining_guesses(self):
    print(f"You have {self.max_guesses - self.num_guesses} guesses left")

  def check_game_over(self):
    self.remaining_guesses()
    return all(letter in self.guesses for letter in self.secret_word), self.max_guesses >= self.num_guesses

limit = LimitedGuessWordGame("word", 5)
limit.play()
limit.make_guess("o")
print(limit.check_game_over())
```

```python
class Flyer:
  def __init__(self, flyer_name : str):
    self.flyer_name = flyer_name

  @staticmethod
  def fly():
    print("I am flying")

class Swimmer:
  def __init__(self, swimmer_name : str):
    self.swimmer_name = swimmer_name

  @staticmethod
  def swim():
    print("I am swimming")

class FlySwimmer(Flyer, Swimmer):
  def __init__(self, flyer_name : str, swimmer_name : str):
    Flyer.__init__(self, flyer_name)
    Swimmer.__init__(self, swimmer_name)

duck = FlySwimmer("The Duck", "swimming")
duck.swim()
duck.fly()
```

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
  def __init__(self, brand : str, speed : int):
    self.brand = brand
    self.speed = speed

  @abstractmethod
  def drive(self):
    pass

class Flying:
  def __init__(self, altitude : int, airport_dest : str):
    self.altitude = altitude
    self.airport_dest = airport_dest

  def flying_where(self):
    return f"Currently flying to {self.airport_dest}"

class FlyingCar(Vehicle, Flying):
  def __init__(self, brand : str, speed : int, altitude : int, airport_dest : str):
    Vehicle.__init__(self, brand, speed)
    Flying.__init__(self, altitude, airport_dest)

  def drive(self):
    return "I am driving"

flyingcar = FlyingCar("The Brand", 70, 80, "TLV")
print(flyingcar.drive())
print(flyingcar.flying_where())

print(flyingcar.brand)
print(flyingcar.speed)
print(flyingcar.altitude)
```

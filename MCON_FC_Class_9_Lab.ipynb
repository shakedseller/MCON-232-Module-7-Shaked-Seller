```python
class Person:
  def __init__(self, name):
    self.name = name

  def speak(self, msg):
    print(msg)

  def walk(self):
    print("I am walking")

# Derived classes

class Student(Person): # student is a person
  def __init__(self, name, age, courses):
    super().__init__("Someone") # brings in all methods from parent
    self.name = name
    self.age = age
    self.courses = courses

  def study(self):
    print("I am studying")

  # override speak
  def speak(self, msg):
    print("Student speak", msg)

  def walk(self):
    print("Student walking")

class Example(Person):
  def __init__(self, name):
    super().__init__(name)

s1 = Student("Duffy", 6, ["CS"])
s2 = Student("Cosmo", 6, ["Composition"])

print(s1.name, s2.name)

s1.speak("something")
s2.speak("mmmmmm")

s1.walk()
s2.walk()
```

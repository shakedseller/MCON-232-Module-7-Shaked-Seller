## Problem 1: Build an Animal Hierarchy

Create a parent class named `Animal`.

### The `Animal` class must:
- store a name
- store an age
- include a method named `speak()` that returns a general animal sound
- include a method named `move()` that returns a general description of movement
- include a method named `describe()` that returns a short sentence about the animal

Then create these child classes:

- `Dog`
- `Cat`
- `Rabbit`

### Child class requirements
Each child class must:
- inherit from `Animal`
- use the parent constructor appropriately
- override `speak()`
- override `move()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `speak()`
- each object gives a different result for `move()`
- `describe()` works for each object


```python
# Write your code for Problem 1 here.
class Animal:
  def __init__(self, name : str, age : int) -> None:
    self.name = name
    self.age = age

  def speak(self):
    return "ah"

  def move(self):
    return "moving"

  def describe(self):
    return "This is a general animal."

class Dog(Animal):
  def __init__(self, name : str, age : int) -> None:
    super().__init__(name, age)

  def speak(self):
    return "bark"

  def move(self):
    return "running"

  def describe(self):
    return "This animal is a dog."

class Cat(Animal):
  def __init__(self, name : str, age : int) -> None:
    super().__init__(name, age)

  def speak(self):
    return "meow"

  def move(self):
    return "gracefully walking"

  def describe(self):
    return "This animal is a cat."

class Rabbit(Animal):
  def __init__(self, name : str, age : int) -> None:
    super().__init__(name, age)

  def speak(self):
    return "(whatever sound rabbits make)"

  def move(self):
    return "hopping"

  def describe(self):
    return "This animal is a rabbit."

d = Dog("The dog", 3)
c = Cat("The cat", 4)
r = Rabbit("The Rabbit", 5)

```

## Problem 2: Build a Vehicle Hierarchy

Create a parent class named `Vehicle`.

### The `Vehicle` class must:
- store a brand
- store a speed
- include a method named `move()` that returns a general statement about transportation
- include a method named `sound_horn()` that returns a general horn sound
- include a method named `describe()` that returns a short sentence about the vehicle

Then create these child classes:

- `Car`
- `Truck`
- `Motorcycle`

### Child class requirements
Each child class must:
- inherit from `Vehicle`
- use the parent constructor appropriately
- override `move()`
- override `sound_horn()`

### After creating your classes
Create one object of each child class and demonstrate that:
- each object stores its own data
- each object gives a different result for `move()`
- each object gives a different result for `sound_horn()`
- `describe()` works for each object


```python
# Write your code for Problem 2 here.
class Vehicle:
  def __init__(self, brand : str, speed : int) -> None:
    self.brand = brand
    self.speed = speed

  def move(self):
    return "The general vehicle is moving"

  def sound_horn(self):
    return "bleep"

  def describe(self):
    return "This is a general vehicle."

class Car(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return "The car drives"

  def sound_horn(self):
    return "beep"

class Truck(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return "The truck drives loudly"

  def sound_horn(self):
    return "HONKKKKKKKK"

class Motorcycle(Vehicle):
  def __init__(self, brand, speed):
    super().__init__(brand, speed)

  def move(self):
    return "The motorcycle strides quickly"

  def sound_horn(self):
    return "boop"

c = Car("The car", 2)
t = Truck("The truck", 3)
m = Motorcycle("The motorcycle", 4)
```


## Problem 3: Demonstrate Polymorphism with Animals

Using your `Dog`, `Cat`, and `Rabbit` classes:

1. Create a list that stores multiple animal objects together.
2. Use a loop to go through the list.
3. For each object, call the **same methods** and show that the results differ depending on the object.

### Required methods to call in the loop
- `speak()`
- `move()`
- `describe()`

## Goal
Your output should clearly demonstrate that the program can treat all of the child objects as `Animal` objects, while each one still behaves in its own way.


```python
# Write your code for Problem 3 here.
animals = [d, c, r]

for a in animals:
  print("----------")
  print(f"{a.name}, {a.age} years old")
  print(a.describe(), a.move(), a.speak(), sep="\n")
```


## Problem 4: Demonstrate Polymorphism with Vehicles

Using your `Car`, `Truck`, and `Motorcycle` classes:

1. Create a list that stores multiple vehicle objects together.
2. Use a loop to go through the list.
3. For each object, call the **same methods** and show that the results differ depending on the object.

### Required methods to call in the loop
- `move()`
- `sound_horn()`
- `describe()`

## Goal
Your output should clearly demonstrate polymorphism in the vehicle family.


```python
# Write your code for Problem 4 here.
vehicles = [c, t, m]

for i in vehicles:
  print("----------")
  print(f"{i.brand}, {i.speed} speed")
  print(i.describe(), i.move(), i.sound_horn(), sep="\n")
```

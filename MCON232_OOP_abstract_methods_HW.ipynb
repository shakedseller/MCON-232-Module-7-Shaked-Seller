# Part 2 — Build an Abstract Parent Class

Create an abstract class called `Animal`.

Requirements:

1. `Animal` must inherit from `ABC`.
2. `Animal` must have a class attribute named `total_animals`, starting at `0`.
3. The constructor must accept `name` and `age`.
4. The constructor must save those as instance attributes.
5. The constructor must increase `total_animals` by 1.
6. Add a concrete method named `show_info()` that returns a string such as `Buddy is 4 years old.`
7. Add a `@classmethod` named `get_total_animals()` that returns the total number of animals created.
8. Add a `@staticmethod` named `is_valid_name(name)` that returns `True` only if `name` is a string and, after stripping spaces, is not empty.
9. Add an abstract method named `speak()`.

Remember:

```python
from abc import ABC, abstractmethod
```

````python
# Part 2 Code Here

from abc import ABC, abstractmethod

# TODO: Build the Animal abstract class here
class Animal(ABC):
  total_animals = 0
  def __init__(self, name : str, age : int) -> None:
    self.name = name
    self.age = age
    Animal.total_animals += 1

  def show_info(self):
    return f"{self.name} is {self.age} years old."

  @classmethod
  def get_total_animals(cls):
    return cls.total_animals

  @staticmethod
  def is_valid_name(name : str) -> bool:
    return True if isinstance(name, str) and len(name.strip()) > 0 else False

  @abstractmethod
  def speak(self):
    pass
````

**Answer:** Every animal makes a different sound, so each child will return something else for speak()

**Answers:**

````python
# Part 7 Fixed Code Here

# TODO: Fix Car so that it follows the abstract contract.
# Then create a list of Vehicle objects and loop through them polymorphically.
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        return f"Brand: {self.brand}"

    @abstractmethod
    def move(self):
        pass

class Car(Vehicle):
    def move(self):
      return "Driving forward"

class Bicycle(Vehicle):
    def move(self):
        return "Pedaling forward"

vehicles = [Car("Toyota"), Bicycle("Trek")]

for v in vehicles:
  print(v.move())
````

# Part 9 — Overriding vs Reassignment vs Shadowing

Read the code carefully, then answer the questions.

````python
class Parent:
    SOUND = "generic sound"

    def __init__(self):
        self.mood = "calm"

    def speak(self):
        return "Parent speaking"

class Child(Parent):
    SOUND = "child sound"

    def __init__(self):
        super().__init__()
        self.mood = "excited"

    def speak(self):
        return "Child speaking"

p = Parent()
c = Child()

print(Parent.SOUND)
print(Child.SOUND)
print(p.mood)
print(c.mood)
print(p.speak())
print(c.speak())
````

### Part 9 Questions

2. Which example shows instance attribute reassignment?
3. Which example shows class attribute shadowing?
4. Why does `Parent.SOUND` still exist even after `Child.SOUND` is defined?
5. Why does `self.mood = "excited"` replace the value for the child object?

**Answers:**
1. The speak() method in the child.
2. self.mood in the child class.
3. The SOUND attribute in the child class.
4. Parent.SOUND is shadowed, so it can still be accessed. It is not destroyed.
5. The parent constructor is called, but then the child replaces "calm" with "excited".

## Part 12 — Multiple Inheritance with a Mixin

In Python, a class can inherit from more than one parent class. This is called **multiple inheritance**.

A common safe use of multiple inheritance is a **mixin**.

A mixin is a small helper class that adds one focused behavior to another class.

In this problem, you will create a mixin called `AdoptableMixin` and use it with one of your `Animal` child classes.

### Important Idea

- `Animal` provides the main identity and required abstract behavior.
- `AdoptableMixin` provides one extra reusable behavior.
- The final child class can inherit from both.

Example structure:

```python
class AdoptableDog(Dog, AdoptableMixin):
    ...
```

---

## Requirements

Create a class named `AdoptableMixin`.

It should have:

- an instance attribute named `adopted`, starting as `False`
- a method named `adopt()` that changes `adopted` to `True`
- a method named `is_adopted()` that returns the value of `adopted`

Then create a class named `AdoptableDog` that inherits from both `Dog` and `AdoptableMixin`.

`AdoptableDog` must:

- inherit from `Dog` and `AdoptableMixin`
- accept `name`, `age`, and `breed`
- call Dog’s constructor correctly
- call AdoptableMixin’s constructor correctly
- still be able to use `show_info()`
- still be able to use `speak()`
- be able to use `adopt()`
- be able to use `is_adopted()`

### Hint

Because both parent classes need setup, your `AdoptableDog` constructor should explicitly call both parent constructors.

```python
Dog.__init__(self, name, age, breed)
AdoptableMixin.__init__(self)
```

This is not the only possible advanced design, but it is the clearest design for this homework.

---

````python
# Part 12 Code Here

# TODO: Create AdoptableMixin
class AdoptableMixin:
  def __init__(self):
    self.adopted = False

  def adopt(self):
    self.adopted = True

  def is_adopted(self):
    return f"Adopted status: {self.adopted}"

# TODO: Create AdoptableDog using multiple inheritance
class AdoptableDog(Dog, AdoptableMixin):
  def __init__(self, name : str, age : int, breed : str) -> None:
    Dog.__init__(self, name, age, breed)
    AdoptableMixin.__init__(self)
# TODO: Create an AdoptableDog object
adoptable_dog = AdoptableDog("The Dog", 1, "Golden Retriever")
# TODO: Print its info and sound
print(adoptable_dog.show_info())
# TODO: Check whether it is adopted
print(adoptable_dog.is_adopted())
# TODO: Call adopt()
adoptable_dog.adopt()
# TODO: Check whether it is adopted again
print(adoptable_dog.is_adopted())
````
